# Tree Data Structure

           |------17-----|
           |             |
    |------4----|  |-----20-----|
    |           |  |            |
    1           9  18          23------|
                                      34

```python
class BinarySearchTreeNode:
    def __init__(self,data):
        self.data=data
        self.right=None
        self.left=None
        
        
    def add_child(self,data):
        if data==self.data:
            return
        
        if data<self.data:
            if self.left:
                self.left.add_child(data)
            else:
                self.left=BinarySearchTreeNode(data)
        else:
            if self.right:
                self.right.add_child(data)
            else:
                self.right=BinarySearchTreeNode(data)
                
     # in order Traversing           
    def in_order_traversal(self):
        
        elements=[]
        # visit left sub tree
        if self.left:
            elements+=self.left.in_order_traversal()
        
        # visit parent
        elements.append(self.data)
        
        # visit right sub tree
        
        if self.right:
            elements+=self.right.in_order_traversal()
        
        return elements
    
    def pre_order_traversing(self):
        elements=[]
        
        if self.data:
            elements.append(self.data)

            
        if self.left:
            elements+=self.left.pre_order_traversing()
            
        if self.right:
            elements+=self.right.pre_order_traversing()
        return elements
    
    def post_order_traversal(self):
        elements=[]
            
        if self.left:
            elements+=self.left.pre_order_traversing()
            
        if self.right:
            elements+=self.right.pre_order_traversing()
            
        if self.data:
            elements.append(self.data)
        return elements
    
    def search(self,val):
        if self.data==val:
            return val
        if self.data>val:
            if self.left:
                if self.left.search(val):
                    return val
                return False
            
        if self.data<val:
            if self.right:
                if self.right.search(val):
                    return val
                return False
            
    def find_max(self):
        if self.right is None:
            return self.data
        return self.right.find_max()
    
    def find_min(self):
        if self.left is None:
            return self.data
        return self.left.find_min()
    
    def total_Sum(self):
        list_of_numbers=self.in_order_traversal()
        return sum(list_of_numbers)
    
def build_tree(nums):
    root=BinarySearchTreeNode(nums[0])
    for n in range(1,len(nums)):
        root.add_child(nums[n])
    print(root.in_order_traversal())
    print(root.pre_order_traversing())
    print(root.post_order_traversal())
    print(root.search(1))
    print("Max Number in tree is:",root.find_max())
    print("Min Number in tree is:",root.find_min())
    print("Sum of all tree nodes are:",root.total_Sum())
    
    return root

if __name__ == '__main__':
    numbers=[17,4,1,20,9,23,18,34]
    build_tree(numbers)
    
            
```
